```python
import pandas as pd
import numpy as np

# Estimación de gastos en Euros
expenses = {
    'Hotel (5 estrellas)': 250 * 7,  # Asumiendo 200 euros por noche
    'Comidas en restaurantes': 150 * 7,  # Asumiendo 150 euros por día
    'Alquiler de coche con conductor': 160 * 7,  # Asumiendo 160 euros por día
    'Espectáculos (museos, teatros)': 300,  # Asumiendo un total de 300 euros
    'Otros gastos (souvenirs, etc.)': 100  # Asumiendo 100 euros en total
}

# Crear un DataFrame con los datos
df = pd.DataFrame(list(expenses.items()), columns=['Gasto', 'Costo'])

# Agregar una nueva columna para indicar si el gasto es computable
non_computable_expenses = ['Otros gastos (souvenirs, etc.)']
df['Es computable'] = ~df['Gasto'].isin(non_computable_expenses)

# Crear un DataFrame para el gasto de 'Copas'
df_copas = pd.DataFrame([['Copas', 18 * 5, False]],
                        columns=['Gasto', 'Costo', 'Es computable'])

# Concatenar los dos DataFrames
df = pd.concat([df, df_copas], ignore_index=True)

# Calcular los totales antes de eliminar la columna 'Es computable'
total_computable = df[df['Es computable']]['Costo'].sum()
total_no_computable = df[~df['Es computable']]['Costo'].sum()

# Eliminar la columna 'Es computable'
df = df.drop(columns='Es computable')

# Reordenar las columnas
df = df[['Costo', 'Gasto']]

# Mostrar la tabla
print(df)

print(f"\nEl total de gastos computables es: {total_computable} Euros")
print(f"El total de gastos no computables es: {total_no_computable} Euros")
```

```python
import matplotlib.pyplot as plt

# Definir los datos
total = 4220
hotel = 1750
non_computable = 160
resto = total - hotel - non_computable

# Crear una lista con los valores
values = [hotel, non_computable, resto]

# Crear una lista con las etiquetas
labels = ['Hotel', 'Gastos no computables', 'Resto']

# Definir una función personalizada para autopct


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} €)".format(pct, absolute)


# Crear el gráfico de tarta
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct=lambda pct: func(pct, values))

# Mostrar el gráfico
plt.show()
```
