```python
import matplotlib.pyplot as plt
import numpy as np

# Diccionario con los datos originales
data = {'Alicante':90, 'Barcelona':113, 'Bilbao':66, 'Canarias':94, 'Cordoba':58, 'Granada':55, 
        'Madrid':83, 'Malaga':100, 'Marbella':204, 'Baleares':124, 'Sevilla':76, 'Valencia':81, 'Zaragoza':43}

cities = list(data.keys())
values = list(data.values())
values_increased = [value * 1.3 for value in values]  # Aumentar los valores en un 30%

# Crear un array con las posiciones de las barras
bar_positions = np.arange(len(cities)) 

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(12, 8))  
width = 0.35  # Ancho de las barras

# Crear las barras para los valores originales
original_bars = ax.bar(bar_positions - width/2, values, width, label='RevPAR original')

# Crear las barras para los valores aumentados
increased_bars = ax.bar(bar_positions + width/2, values_increased, width, label='RevPAR con IA')

# Añadir etiquetas, título y leyenda
ax.set_ylabel('Valores')
ax.set_title('Valores por ciudad')
ax.set_xticks(bar_positions)
ax.set_xticklabels(cities)
ax.legend()

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation='vertical')

# Mostrar el gráfico
plt.show()
```
